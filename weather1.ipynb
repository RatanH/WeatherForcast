```python
import pandas as pd

weather = pd.read_csv("./data/kaggle/IndianWeatherRepository.csv",index_col='last_updated')
```

```python
weather
```

```python
#read columns
weather.columns
```

### Preprocessing

```python
#checking datatypes
weather_obj = weather.select_dtypes(include='object')
weather_obj.head()
```

```python
#creating a series of col names having dtype=object
obj_cols = pd.Series(weather_obj.columns)
obj_cols
```

```python
#copy of the weather data
weather_num = weather.copy()
#remove object columns, leaving only numeric cols for analysis
weather_num.drop(columns=obj_cols,axis=1,inplace=True)
weather_num.columns
```

```python
#Removing redundant columns
weather_num.drop(['temperature_fahrenheit','wind_mph','pressure_in','precip_mm','feels_like_fahrenheit','visibility_km','gust_mph'], axis=1,inplace=True)
weather_num.columns
```

```python
#check for nulls
weather_num.isnull().sum()
```

```python
#initial visual analysis (histograms) 
weather_num.hist(bins=16,figsize=(20,15))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Creating a correlation matrix
sns.set(font_scale=0.9)

weather_corr = weather_num.corr(method="kendall")

plt.figure(figsize=(20, 15))
heatmap= sns.heatmap(weather_corr, vmin=-1, vmax=1, annot=True,cmap='BrBG',annot_kws={"fontsize":4},linewidths=0.1)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':2}, pad=12);
```
